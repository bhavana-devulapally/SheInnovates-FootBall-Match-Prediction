```python
#Convert the data to proper date time format.
matches_data["date"] = pd.to_datetime(matches_data["date"])

```

```python
#Create a new column target nad if the team has won the matched it enters 1 or else 0 to all other possibilities.
matches_data["target"]= (matches_data["result"]=="W").astype("int")
matches_data
```

```python
#If match is played inthe home state it is assigned to one or else 0
matches_data["venue_code"] = matches_data["venue"].astype("category").cat.codes
```

```python
#Assign random numbers to each opponet team in the opp_code column
matches_data["opp_code"]=matches_data["opponent"].astype("category").cat.codes
```

```python
#Just take the hour value and print it in the column
matches_data["hour"] = matches_data["time"].str.replace(":.+", "", regex=True).astype("int")
matches_data
```

```python
"Map each day of the week to a number 1 means sunday and so on ...."
matches_data["day_code"]=matches_data["date"].dt.dayofweek
matches_data
```

By Using the Random Forest Classifier we now start initializing the model. Random Forest is a ML Model that picks up non linearities in the data.

The data in the above table is not linear always. For example the opp code is not linear which means if there is high opp_code it means its more difficult to defeat that team. Its just random number given to the column opp_code.

For such data, a random model and pick up that sort of data which is not linear as well and train and work it such non-linear data.

Randor forest can deal with such data which is non-linear but a linear model cannot.

Now, we initialize a random classifier class .

```python
#Train the ML Model,
from sklearn.ensemble import RandomForestClassifier
#n_estimators = number of individual descion trees we want ot train. RF is basically a series of desicion trees,each desicion tree having differnet parameters.In other words higher the
#n_estimators is more time algorithm will take to run and it will be more acurate.

#min_samples_split is the number of samples we want in the  leaves in a desicion tree. That means higher the value, less likely to overfitting and accuracy is less.
#Random state means if we run the random forst model multiple times we would be getting the same results rather than change of result everytime which causes confusion.as long as data is same.
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
#Splitting of Training and Testing data.
#Ths is a time sensitive segregation of training and testing data which means, we have be sure the testing data comes after the training data.while splitting.
#In real time, we train the data only on the past data and not the future data and hence data in the test set comes after the data in the training set.
#Use train the algorithm by using the past data , and get the future results.

#Just write the conditions for training and testing.
train = matches_data[matches_data["date"] < '2022-01-01']
test = matches_data[matches_data["date"] > '2022-01-01']

#X
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
#Now we fit the RF Model
#Train predictors and test the target. By training thse predictors we cantest the outputs if they win or loose the match.
rf.fit(train[predictors], train["target"])

#Now we can generate predictions using the predict method. Pass in our test data withÂ predictors.
preds = rf.predict(test[predictors])
```

```python
#Now we see the accuracy of the model.
from sklearn.metrics import accuracy_score

#Determine the accuracy of the model.
#1. Import accuracy_score. Accuracy_score is a metric that will say if you predicted a win, what percentage of the times your predictionÂ accurate.
acc = accuracy_score(test["target"], preds)
acc
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group

cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches_data.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))

matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error

combined, error = make_predictions(matches_rolling, predictors + new_cols)
error
```
